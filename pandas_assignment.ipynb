```python
#Identifying outliers using iqr 
Q1=df['Runtime (Minutes)'].quantile(0.25)
Q3=df['Runtime (Minutes)'].quantile(0.75)
```

```python
Q1,Q3
```

```python
IQR=Q3-Q1
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

```python
outliers=df[(df['Runtime (Minutes)'] > lower_limit) & (df['Runtime (Minutes)'] < upper_limit)]
```

```python
#Handling the outliers
filtered_data=df[(df['Runtime (Minutes)'] >= lower_limit) & (df['Runtime (Minutes)'] <= upper_limit)]
```

```python
filtered_data
```

```python
plt.figure(figsize=(10,8))
plt.boxplot(filtered_data['Runtime (Minutes)'])
plt.grid()
plt.show()
```
