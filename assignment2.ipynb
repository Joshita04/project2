```python

df['Mileage_numerical'] = df['Mileage'].str.extract('([\d.]+)').astype(float)

df
```

```python

brand_groups = df.groupby('Name')
brand_groups
```

```python
brand_correlations = brand_groups[['Mileage_numerical', 'Price']].corr()
```

```python
import matplotlib.pyplot as plt

for brand, data in brand_groups:
    plt.scatter(data['Mileage_numerical'], data['Price'], label=brand,color='red')
    plt.xlabel('Mileage (kmpl)')
    plt.ylabel('Price (in lakhs)')
    plt.legend()
    plt.title(f'Mileage vs. Price for {brand}')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression 

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error




df['Brand'] = df['Name'].apply(lambda x: x.split()[0])


X = df[['Mileage_numerical']]
y = df['Price']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


model = LinearRegression()

model.fit(X_train, y_train)


y_pred = model.predict(X_test)


r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

result = pd.DataFrame({
    'Mileage_numerical': X_test['Mileage_numerical'],
    'Actual_Price': y_test,
    'Predicted_Price': y_pred,
})
result['Error'] = abs(result['Actual_Price'] - result['Predicted_Price'])
result['Error_Percentage'] = (result['Error'] / result['Actual_Price']) * 100




result

```
